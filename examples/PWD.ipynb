```julia
using PyPlot,Seismic
```

```julia
download("http://ualberta.ca/~kstanton/files/kevin_0off_velocity.su","kevin_0off_velocity.su");
SegyToSeis("kevin_0off_velocity.su","kevins_model",format="su",swap_bytes=false);
```

```julia
d,h,ext = SeisRead("kevins_model");
param = Dict(:w1=>30,:w2=>20)#,:dx_in=>8,:dz_in=>4);
p=0
coh,pp,res = SeisPWD(d; param...);
```

```julia
SeisPlot(d,cmap="gray"); 
```
